*Ejercicio 2: Dado el diccionario*

```python
datos = {'a':10, 'b':25, 'c':7, 'd':18}

# a) Claves con valor mayor a 12
claves_mayores = [k for k, v in datos.items() if v > 12]
print(claves_mayores)

# b) Nuevo diccionario con valores al cuadrado
dict_cuadrado = {k: v**2 for k, v in datos.items() if v > 12}
print(dict_cuadrado)
```

*Ejercicio 3: Define una función 'resumen(lista)' que retorne un diccionario con:longitud, suma, promedio, mínimo y máximo de la lista.*

```python
def resumen(lista):
    return {
        'longitud': len(lista),
        'suma': sum(lista),
        'promedio': sum(lista) / len(lista),
        'minimo': min(lista),
        'maximo': max(lista)
    }

print(resumen([3, 1, 7, 2, 5]))
```

*Ejercicio 4: Usa map() para aplicar la función potencia(x, 3) a la lista [1,2,3,4,5].*

```python
def potencia(x, exp=3):
    return x ** exp

lista = [1, 2, 3, 4, 5]
resultado = list(map(lambda x: potencia(x), lista))
print(resultado)
```

*Ejercicio 1: Dado A = np.arange(1, 26).reshape(5, 5)*

```python
A = np.arange(1, 26).reshape(5, 5)
print(A)

# a) Sub-matriz: 3 ultimas filas, 3 primeras columnas
sub = A[-3:, :3]
print('Sub-matriz:\n', sub)

# b) Segunda columna completa
col2 = A[:, 1]
print('Segunda columna:', col2)

# c) Elementos de las esquinas
esquinas = A[[0, 0, -1, -1], [0, -1, 0, -1]]
print('Esquinas:', esquinas)
```

```python
# a) Mascara para valores mayores a 12
mascara = A > 12
print('Mascara:\n', mascara)

# b) Reemplazar por -1 en copia
B = A.copy()
B[mascara] = -1
print('Copia con -1:\n', B)

# c) np.where: impares por 0, pares se mantienen
C = np.where(A % 2 == 1, 0, A)
print('Impares a 0:\n', C)
```

*Ejercicio 1: Crea una matriz 5×4 de enteros aleatorios entre 1 y 100 (seed=7).*

```python
#Fórmula: (x - min_col) / (max_col - min_col)
np.random.seed(7)
M = np.random.randint(1, 101, size=(5, 4))
print(M)

# a) Media, max, min por columna
print('Media columnas:', M.mean(axis=0))
print('Max columnas:', M.max(axis=0))
print('Min columnas:', M.min(axis=0))

# b) Suma acumulada primera fila
print('Suma acumulada fila 0:', np.cumsum(M[0]))

# c) Normalizar columnas entre 0 y 1
min_col = M.min(axis=0)
max_col = M.max(axis=0)
normalizada = (M - min_col) / (max_col - min_col)
print('Normalizada:\n', normalizada)
```

*Ejercicio 4: Crea dos matrices 3×4 y concaténalas vertical y horizontalmente.*

```python
#Luego divide el resultado vertical en 3 partes iguales con vsplit.
A = np.arange(12).reshape(3, 4)
B = np.arange(12, 24).reshape(3, 4)

# Concatenar vertical
vertical = np.vstack((A, B))
print('Vertical 6x4:\n', vertical)

# Concatenar horizontal
horizontal = np.hstack((A, B))
print('Horizontal 3x8:\n', horizontal)

# Dividir vertical en 3 partes
partes = np.vsplit(vertical, 3)
print('Partes:', len(partes))
for i, p in enumerate(partes):
    print(f'Parte {i} shape:', p.shape)
```
